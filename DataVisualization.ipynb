### LINEPLOT

- A line plot is a graphical representation of data that displays data points along a number line. 
- Line plots are commonly used in statistics and data analysis to visualize the relationship between two numerical variables.


```python
sns.get_dataset_names()
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
#Functional magnetic resonance imaging, or fMRI, is a technique for measuring brain activity. 
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
fmri.shape
```

```python
event_values=fmri['event'].unique()
event_values
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event", style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event", style="event", markers=True)
plt.show()
```

### BARPLOT
- Bar plots are commonly used to visualize data that has been organized into categories, such as the results of a survey.

### SCATTERPLOT
- A scatter plot is a graphical representation of data that uses dots or other symbols to show the relationship between two numerical variables.

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris["species"].unique()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris)
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, hue="species")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, hue="petal_length")
plt.show()
```

### HISTOGRAM
- Histograms are commonly used to visualize data that is distributed over a continuous range.
- For e.g., distribution of temperatures over the course of a year

```python
diamonds=sns.load_dataset("diamonds")
diamonds.head()
```

```python
diamonds["cut"].unique()
```

```python
sns.histplot(diamonds["price"]) #displot #distplot
plt.show()
```

```python
sns.distplot(diamonds["price"]) #displot
plt.show()
```

```python
sns.histplot(diamonds["price"], color="green") #displot
plt.show()
```

```python
sns.histplot(diamonds["price"],kde=True, bins=5) #displot
plt.show()
```

### JOINT PLOT
- A joint plot, also known as a joint distribution plot, is a graphical representation of data that shows the relationship between two numerical variables.
- It represents the distribution of each variable and also the correlation between the variables

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris)
plt.show()
```

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris, color="orange", kind="reg")
plt.show()
```

### HEATMAP
- A heatmap is a graphical representation of data where the values of the data are represented as colors. 
-  Heatmaps can also be used to compare multiple sets of data, or to identify clusters or groups within the data

```python
glue = sns.load_dataset("glue").pivot("Model", "Task", "Score")
glue.head()
```

```python
sns.heatmap(glue, annot=True, linewidths=1, cmap="Greens")
```

### Pie Plot
- Pie plots are useful for showing the relative proportions of the different data points in a dataset.

```python
# declaring data
data = [44, 45, 40, 41, 39]
keys = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E']
  
explode = [0, 0.1, 0, 0, 0]
palette_color = sns.color_palette('bright')
  
plt.pie(data, labels=keys, colors=palette_color,
        explode=explode, autopct='%.0f%%')

plt.show()
```

### Subplots

```python
import numpy as np
  
# making subplots objects
fig, ax = plt.subplots(2, 2)
  
# create data
x = np.linspace(0, 10, 1000)
  
# draw graph
ax[0, 1].plot(x, np.sin(x), 'r-.')
ax[0, 0].plot(x, np.cos(x), 'g--')
ax[1, 0].plot(x, np.tan(x), 'y-')
ax[1, 1].plot(x, np.sinc(x), 'c.-')
  
plt.show()
```
